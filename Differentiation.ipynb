1. Вывести формулы численного дифференцирования для вычисления первой производной (правая и центральная разности) и второй производной (симметричные формулы второго и четвертого порядка).

*Постановка задачи:*<br>Пусть задана непрерывная функция $f(x)$ на $[a;b]$ <br>
и равномерная сетка значений ${x}^N_{i=0}$ на $[a; b]$, для которого $x_0=a$ и $x_N =b$<br>Из математического анализа формула производной через предел $f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}$<br>Пусть $h$ - шаг сетки, где $h = x_{i+1} - x_i$<br><br>*Вывод правой разности*:<br>Способ 1:<br>Зададим множество значений функции $F(i)$ из непрерывно заданной функции $f(x)$ соотвественно как $F(i) = f(x_i)$ где $i = \overline{0, N}$<br>  Тогда получим правую разность из формулы $F'(i) = \lim_{h \to 0} \frac{F(i + 1) - F(i)}{h}$<br>Ввиду малости $h$ можно приблизительно утвержать следующую формулу:<br>
$F'(i) \approx \frac{F(i + 1) - F(i)}{h}$<br>Способ 2 (Через Тейлора):

$
F(i+1) = F(i) + h F'(i) + \mathcal{O}(h^2)$
<br>
Пренебрегаем $\mathcal{O}(h^2)$ и получим путем преобразований:<br>
$F'(i) \approx \frac{F(i + 1) - F(i)}{h}$

<br>*Вывод центральной разности*:<br>Способ 1:<BR>Зададим две точки $c_+\in[x_i, x_{i+1}]$ и $c_-\in[x_{i-1}, x_i]$<br>По приблизительной правой разности для этих точек можно утверждать $f'(c_+) \approx \frac{F(i + 1) - F(i)}{h}$ и $f'(c_-) \approx \frac{F(i) - F(i - 1)}{h}$<br> тогда получим более точную формулу для нахождения $F'(i) \approx \frac{f'(c_-) + f'(c_+)}{2}$<br> отсюда получим формулу центральной разности:<br>$F'(i) \approx \frac{F(i+1) - F(i-1)}{2h}$<br>
Способ 2 (Через Тейлора):<br>
$
F(i+1) = F(i) + h F'(i) + \frac{h^2}{2} F''(i) + +\mathcal{O}(h^3)$<br>
$
F(i-1) = F(i) - h F'(i) + \frac{h^2}{2} F''(i) +\mathcal{O}(h^3)$<br>
Путем создания линейной комбинации избавимся от членов нулевой и второй производной<br>

$F(i+1) - F(i-1) = 2hF'(i) + \mathcal{O}(h^3)$<BR>
Пренебрегаем $\mathcal{O}(h^3)$ и получим путем преобразований:<br>
$F'(i) \approx \frac{F(i+1) - F(i-1)}{2h}$

<br>*Вывод симметричной формулы второго порядка*:<br>Способ 1:
<br>Воспользуемся правой разностью для решения задачи соотственно<br> $F'(i) = \frac{F(i + 1) - F(i)}{h}$<br>$F'(i - 1) = \frac{F(i) - F(i - 1)}{h}$<br>Из формулы левой разности для второго порядка получим:<br>$F''(i) = \frac{F'(i) - F'(i - 1)}{h}$ откуда сопоставим формулы и получим итоговую:<br>$F''(i) \approx \frac{F(i+1) - 2F(i) + F(i-1)}{h^2}$<br>
2 Способ (Через Тейлора):<br>
Разложение в ряд Тейлора вокруг точки $x_i$:<br>
$F(i + 1) = F(i) + hF'(i) + \frac{h^2}{2}F''(i) + \frac{h^3}{6}F'''(i) + \mathcal{O}(h^4)$<br>
$F(i - 1) = F(i) - hF'(i) + \frac{h^2}{2}F''(i) - \frac{h^3}{6}F'''(i) + \mathcal{O}(h^4)$<br>
Путем создания линейной комбинации избавимся от членов первой и третьей производной и получим:<br>
$F(i+1) + F(i-1) = 2 F(i) + h^2F''(i) + \mathcal{O}(h^4)$<BR>
Пренебрегай $\mathcal{O}(h^4)$ получим итоговую формулу:<BR>
$F''(i) \approx \frac{F(i+1) - 2F(i) + F(i-1)}{h^2}$

2.1 Реализация функции diff_1_right и diff_1_central, вычисляющая первую производную с использованием правой и центральной разности

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def function(x: float) -> float:
    return math.cos(x) * math.cos(x)

def function_np(x: np.ndarray) -> np.ndarray:
    return np.cos(x) * np.cos(x)

def derived_function_np(x: np.ndarray) -> np.ndarray:
    return -2 * np.cos(x) * np.sin(x)

def derived_function(x: float) -> float:
    return -2 * math.cos(x) * math.sin(x)

def diff_1_right(x_i: float, h: float) -> float:
    return (function(x_i + h) - function(x_i)) / h

def diff_1_central(x_i: float, h: float) -> float:
    return (function(x_i + h) - function(x_i - h)) / (2 * h)

def diff_1_right_np(x: np.ndarray, h: float) -> np.ndarray:
    return (function_np(x + h) - function_np(x)) / h

def diff_1_central_np(x: np.ndarray, h: float) -> np.ndarray:
    return (function_np(x + h) - function_np(x - h)) / (2 * h)
```

2.2 Используя реализованные функции, вычислить производные от сеточной функции $F(i)$, полученные вычислением $f(x)$ на достаточно подробной сетке. Сравнить визаульно полученный результат с реальным значением производной

```python
# Параметры сетки
a = -1.5
b = 1.5
N = 15
h = (b - a) / (N - 1)
x_array = [a + i * h for i in range(N)]
x_smooth = np.linspace(a, b, 10000)

# Вычисление производных
F_array_right = [diff_1_right(x_cur, h) for x_cur in x_array]
F_array_central = [diff_1_central(x_cur, h) for x_cur in x_array]

# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(x_array, F_array_right, label='Правая разность', color="red")
plt.plot(x_array, F_array_central, label='Центральная разность', color='green')
plt.plot(x_smooth, derived_function_np(x_smooth), label='Точная производная', color='blue')

# Настройка графика
plt.xlabel('x')
plt.ylabel("y'")
plt.title('Сравнение производных')
plt.legend()
plt.grid(True)
plt.show()
```

2.3 На одном графике построить зависимость погрешности от шага сетки в логарифмическом масштабе для реализованных функций. Определить из графиков порядок точности численного дифференцированя для каждого метода. Вычислить оценку оптимального шага сетки. Изобразить на графике в виде вертикальной линии. По графику определить фактический оптимальный шаг сетки.

```python
def chebyshev_norm(approx: list[float], exact: list[float]) -> float:
    return max(abs(a - b) for a, b in zip(approx, exact))

def chebyshev_norm_np(approx: np.ndarray, exact: np.ndarray) -> np.ndarray:
    return np.max(np.abs(approx - exact))

h_list = []
error_right = []
error_central = []
r = 1.5           # коэффициент уменьшения шага

h_start = 1
h_end = 1e-7/8

h = h_start
while h >= h_end:
    x_array_np = np.arange(a, b + h, h)

    F_array_right = diff_1_right_np(x_array_np, h)
    F_array_central = diff_1_central_np(x_array_np, h)
    F_array_exact = derived_function_np(x_array_np)

    error_right.append(chebyshev_norm_np(F_array_right, F_array_exact))
    error_central.append(chebyshev_norm_np(F_array_central, F_array_exact))
    h_list.append(h)

    h /= r

# Построение графика
plt.figure(figsize=(10, 6))
plt.loglog(h_list, error_right, color="red", label='Правая разность')
plt.loglog(h_list, error_central, color="green", label='Центральная разность')


h_opt_right = 1e-7/8
h_opt_central = 4.64e-06

# Изображение оптимальных шагов на графике
plt.axvline(h_opt_right, color='red', linestyle='--', label=f'Оптимальный шаг (правая разность): {h_opt_right:.2e}')
plt.axvline(h_opt_central, color='green', linestyle='--', label=f'Оптимальный шаг (центральная разность): {h_opt_central:.2e}')

# Настройка графика
plt.xlabel('Шаг сетки (h)')
plt.ylabel('Погрешность (Чебышевская норма)')
plt.title('Зависимость погрешности от шага сетки')
plt.legend()
plt.grid(True)
plt.show()
```

3.1 Реализация функции diff_2_ord2 и diff_2_ord4 для численного вычисления второй производной методом центральных разностей со вторым и четвертым порядком точности

```python
def derived_derived_function(x: float) -> float:
    return 2 *(math.sin(x) * math.sin(x) - math.cos(x) * math.cos(x))

def derived_derived_function_np(x: np.ndarray) -> np.ndarray:
    return 2 *(np.sin(x) * np.sin(x) - np.cos(x) * np.cos(x))

def diff_2_ord2(x_i: float, h: float) -> float:
    return (function(x_i + h) - 2* function(x_i) + function(x_i - h)) / (h * h)

def diff_2_ord4(x_i: float, h: float) -> float:
    return (- function(x_i + 2 * h) + 16 * function(x_i + h) - 30 * function(x_i) + 16 * function(x_i - h) - function(x_i - 2 * h)) / (12 * h**2)
```

3.2 Используя реализованные функции, вычислить вторые производные от сеточной функции  F(i) , полученные вычислением  f(x)  на достаточно подробной сетке. Сравнить визаульно полученный результат с реальным значением второй производной

```python
# Параметры сетки
a = -1.5
b = 1.5
N = 10
h = (b - a) / (N - 1)
x_array = [a + i * h for i in range(N)]
x_smooth = np.linspace(a, b, 10000)

# Вычисление вторых производных
F_array_right = [diff_2_ord2(x_cur, h) for x_cur in x_array]
F_array_central = [diff_2_ord4(x_cur, h) for x_cur in x_array]

# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(x_array, F_array_right, label='Центральная разность второго порядка второй производной', color="red")
plt.plot(x_array, F_array_central, label='Центральная разность четвертого порядка второй производной', color='green')
plt.plot(x_smooth, derived_derived_function_np(x_smooth), label='Точная вторая производная', color='blue')

# Настройка графика
plt.xlabel('x')
plt.ylabel("y'")
plt.title('Сравнение вторых производных')
plt.legend()
plt.grid(True)
plt.show()
```

3.3 На одном графике построить зависимость погрешности от шага сетки в логарифмическом масштабе для реализованных функций. Определить из графиков порядок точности численного дифференцированя для каждого метода. Вычислить оценку оптимального шага сетки. Изобразить на графике в виде вертикальной линии. По графику определить фактический оптимальный шаг сетки.

```python
h_list = []
error_2 = []
error_4 = []
r = 1.5           # коэффициент уменьшения шага

h_start = 1
h_end = 1e-6

h = h_start
while h >= h_end:
    x_array = [a + i * h for i in range(int((b - a) / h) + 1)]

    F_array_2 = [diff_2_ord2(x_cur, h) for x_cur in x_array]
    F_array_4 = [diff_2_ord4(x_cur, h) for x_cur in x_array]
    F_array_exact = [derived_derived_function(x_cur) for x_cur in x_array]

    error_2.append(chebyshev_norm(F_array_2, F_array_exact))
    error_4.append(chebyshev_norm(F_array_4, F_array_exact))
    h_list.append(h)

    h /= r

# Построение графика
plt.figure(figsize=(10, 6))
plt.loglog(h_list, error_2, color="red", label='Центральная разность второй производной второго порядка')
plt.loglog(h_list, error_4, color="green", label='Центральная разность второй производной четвертого порядка')

h_opt_2 = 1.58e-04
h_opt_4 = 3.31e-03

# Изображение оптимальных шагов на графике
plt.axvline(h_opt_2, color='red', linestyle='--', label=f'Оптимальный шаг (второй порядок): {h_opt_2:.2e}')
plt.axvline(h_opt_4, color='green', linestyle='--', label=f'Оптимальный шаг (четвертый порядок): {h_opt_4:.2e}')

# Настройка графика
plt.xlabel('Шаг сетки (h)')
plt.ylabel('Погрешность (Чебышевская норма)')
plt.title('Зависимость погрешности от шага сетки')
plt.legend()
plt.grid(True)
plt.show()
```
